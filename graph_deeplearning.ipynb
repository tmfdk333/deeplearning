```python
import numpy as np
```

```python
import matplotlib.pylab as plt
```

## **AND**

```python
plt.plot([0,1,0], [0,0,1], 'bv')
plt.plot([1],[1], 'ro')
plt.axis([-0.5, 1.5, -0.5, 1.5])
x = np.arange(-0.8, 1.8, 0.1)
y = np.arange(-0.8, 1.8, 0.1)
plt.plot(x, x*0, linestyle='--', color="gray", linewidth=1)
plt.plot(0*y, y, linestyle='--', color="gray", linewidth=1)
plt.plot(x, x*-1 + 1.5)
plt.ylabel("x2")
plt.xlabel("x1")
plt.title("AND")
plt.show()
```

## **NAND**

```python
plt.plot([0,1,0], [0,0,1], 'ro')
plt.plot([1],[1], 'bv')
plt.axis([-0.5, 1.5, -0.5, 1.5])
x = np.arange(-0.8, 1.8, 0.1)
y = np.arange(-0.8, 1.8, 0.1)
plt.plot(x, x*0, linestyle='--', color="gray", linewidth=1)
plt.plot(0*y, y, linestyle='--', color="gray", linewidth=1)
plt.plot(x, x*-1 + 1.5)
plt.ylabel("x2")
plt.xlabel("x1")
plt.title("NAND")
plt.show()
```

## **OR**

```python
plt.plot([0], [0], 'bv')
plt.plot([1, 0, 1],[0, 1, 1], 'ro')
plt.axis([-0.5, 1.5, -0.5, 1.5])
x = np.arange(-0.8, 1.8, 0.1)
y = np.arange(-0.8, 1.8, 0.1)
plt.plot(x, x*0, linestyle='--', color="gray", linewidth=1)
plt.plot(0*y, y, linestyle='--', color="gray", linewidth=1)
plt.plot(x, x*-1 + 0.5)
plt.ylabel("x2")
plt.xlabel("x1")
plt.title("OR")
plt.show()
```

## **XOR**

```python
plt.plot([0, 1],[0, 1], 'bv')
plt.plot([1, 0],[0, 1], 'ro')
plt.axis([-0.5, 1.5, -0.5, 1.5])
x = np.arange(-0.8, 1.8, 0.1)
y = np.arange(-0.8, 1.8, 0.1)
plt.plot(x, x*0, linestyle='--', color="gray", linewidth=1)
plt.plot(0*y, y, linestyle='--', color="gray", linewidth=1)
plt.plot(x, x*-1 + 0.5)
plt.plot(x, x*-1 + 1.5)
plt.ylabel("x2")
plt.xlabel("x1")
plt.title("XOR")
plt.show()
```

## **Step Function**

```python
def step_function(x):
    return np.array(x>0, dtype=np.int)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
```

```python
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## **Sigmoid**

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## **Step Function & Sigmoid**

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = step_function(x)
y2 = sigmoid(x)
plt.plot(x, y1, linestyle="--")
plt.plot(x, y2)
plt.ylim(-0.1, 1.1)
plt.show()
```

## **ReLU**

```python
#http://quiet-time.tistory.com/10
def relu(x): 
    return np.maximum(0, x)

x = np.arange(-6.0, 6.0, 0.1)
y= relu(x)
#plt.figure(figsize=(10,10))    
plt.plot(x,y)
plt.show()
```
